# Exercise on function

```python
#Unit Conversion Program
def convert_units(conversion_type, value):
    if conversion_type == 1:
        return value * 3.28084
    elif conversion_type == 2:
        return value / 3.28084
    elif conversion_type == 3:
        return value * 2.20462
    elif conversion_type == 4:
        return value / 2.20462
    elif conversion_type == 5:
        return value * 0.264172
    elif conversion_type == 6:
        return value / 0.264172

print("1: meters->feet  \n2: feet->meters")
print("3: kg->lbs      \n4: lbs->kg")
print("5: liters->gallons \n6: gallons->liters")

try:
    c = int(input("Enter conversion type: "))
    v = float(input("Enter value: "))
    print("Converted:", convert_units(c,v))
except:
    print("Invalid input")
```

```python
# Sum, Average, Max, Min from user list
def find_sum(nums):
    return sum(nums)

def find_avg(nums):
    return sum(nums)/len(nums)

def find_max(nums):
    return max(nums)

def find_min(nums):
    return min(nums)

try:
    numbers=input("Enter numbers separated by space: ")
    nums=[float(x) for x in numbers.split()]
    print("Sum:",find_sum(nums))
    print("Average:",find_avg(nums))
    print("Max:",find_max(nums))
    print("Min:",find_min(nums))
except:
    print("Invalid input")
```

# Exercise on Nested List

```python
# Flatten Nested List
def flatten(lst):
    r=[]
    for i in lst:
        if isinstance(i,list):
            r+=flatten(i)
        else:
            r.append(i)
    return r

print(flatten([[1,2],[3,4],[5]]))
```

```python
# Access nested element
lst=[[1,2,3],[4,5,6],[7,8,9]]
print(lst[1][2])
```

```python
#Sum nested list
def sum_nested(lst):
    s=0
    for i in lst:
        if isinstance(i,list):
            s+=sum_nested(i)
        else:
            s+=i
    return s

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
#Remove element
def remove_element(lst,val):
    r=[]
    for i in lst:
        if isinstance(i,list):
            r.append(remove_element(i,val))
        elif i!=val:
            r.append(i)
    return r

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
# Max in nested list
def find_max(lst):
    m=float('-inf')
    for i in lst:
        if isinstance(i,list):
            m=max(m,find_max(i))
        else:
            m=max(m,i)
    return m

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
# Count occurrences
def count_occ(lst,val):
    c=0
    for i in lst:
        if isinstance(i,list):
            c+=count_occ(i,val)
        elif i==val:
            c+=1
    return c

print(count_occ([[1,2],[2,3],[2,4]],2))
```

```python
# Deep flatten
def deep_flatten(lst):
    r=[]
    for i in lst:
        if isinstance(i,list):
            r+=deep_flatten(i)
        else:
            r.append(i)
    return r

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
#Average nested list
def flatten(lst):
    r=[]
    for i in lst:
        if isinstance(i,list):
            r+=flatten(i)
        else:
            r.append(i)
    return r

f=flatten([[1,2],[3,4],[5,6]])
print(sum(f)/len(f))
```

# Matrix Operation

```python
#Matrix operations
A=np.array([[3,4],[7,8]])
B=np.array([[5,3],[2,1]])
```

```python
#Prove A * A^-1 = I

A_inv = np.linalg.inv(A)

print("\nInverse of A:")
print(A_inv)

identity_check = np.dot(A, A_inv)

print("\nA * A^-1 :")
print(identity_check)
```

```python
#Prove AB ≠ BA


AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nA * B =")
print(AB)

print("\nB * A =")
print(BA)

print("\nAre AB and BA equal?")
print(np.array_equal(AB, BA))
```

```python
#Prove (AB)^T = B^T A^T


AB_T = (AB).T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T =")
print(AB_T)

print("\nB^T * A^T =")
print(BT_AT)

print("\nAre they equal?")
print(np.array_equal(AB_T, BT_AT))
```

```python
# Solve Linear Equations using Matrix Method
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

B_eq = np.array([[-1],
                 [-3],
                 [9]])

print("\nCoefficient Matrix A:")
print(A_eq)

print("\nConstant Matrix B:")
print(B_eq)
```

```python
# Solve using Inverse Method
A_eq_inv = np.linalg.inv(A_eq)

X = np.dot(A_eq_inv, B_eq)

print("\nSolution using Inverse Method (X = A^-1 B):")
print("x =", X[0][0])
print("y =", X[1][0])
print("z =", X[2][0])
```

```python
# Solve directly using np.linalg.inv()
X2 = np.linalg.inv(A_eq).dot(B_eq)

print("\nSolution using np.linalg.inv function:")
print("x =", X2[0][0])
print("y =", X2[1][0])
print("z =", X2[2][0])
```
